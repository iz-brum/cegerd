```python
import folium
import branca.colormap
import numpy as np
from folium.plugins import HeatMap
import json  # Adicione esta linha para importar o m√≥dulo json
from folium.plugins import minimap
```

DADOS DE ENTRADA

```python
geojson_arquivo = "mt.json"
dados = np.genfromtxt("chuvas.csv", delimiter = ",")
# Seus dados
dados = np.genfromtxt("chuvas.csv", delimiter=",")

# Remover linhas com valores NaN
dados = dados[~np.isnan(dados).any(axis=1)]
```

MAPA

```python

# Criar o mapa de Mato Grosso nas coordenadas aproximadas do centro do estado
mapa_mt = folium.Map([-12.6819, -56.9211], zoom_start=7, tiles=None)

# Adicionar minimapa
minimapa = folium.plugins.MiniMap(tile_layer="cartodbpositron", position="bottomleft")
mapa_mt.add_child(minimapa)


# Adicionar o fundo branco
folium.TileLayer(tiles=branca.utilities.image_to_url([[1, 1], [1, 1]]),
                 attr="Defesa Civil", name="Municipios").add_to(mapa_mt)

# Adicionando a fronteira dos munic√≠pios
estilo = lambda x: {"color": "black", "fillOpacity": 0, "weight": 1}

with open(geojson_arquivo, encoding='utf-8') as f:
    geojson_data = json.load(f)

folium.GeoJson(geojson_data, style_function=estilo, name="Municipios").add_to(mapa_mt)

# Adicionando GeoJson ao mapa com nomes dos munic√≠pios
folium.GeoJson(
    geojson_data,
    style_function=estilo,
    name="Municipios",
    tooltip=folium.GeoJsonTooltip(fields=['name'], labels=True)
).add_to(mapa_mt)

#Paleta de cores
indices = [0, 0.3, 0.7, 1]

colormap = branca.colormap.StepColormap(["green", "yellow", "red"], index = indices,
                                       caption = "√çndice em mm")

dicionario_cores = {0: "green",
                   0.3: "green",
                   0.301: "yellow",
                   0.7: "yellow",
                   0.701: "red",
                   1: "red"}


colormap.scale(0, 70).add_to(mapa_mt)

#Mapa de Calor
HeatMap(data = dados,
       gradient = dicionario_cores,
       min_opacity = 0.1,
       radius = 20,
       blur = 5,
       name = "chuvas.csv").add_to(mapa_mt)


#Controle de camadas
folium.LayerControl(position = "topleft").add_to(mapa_mt)

#Salvando o mapa
mapa_mt.save("mapa de chuvas.html")
mapa_mt

```
